```python
import numpy as np                  # linspace生成固定范围的数组(等差数列函数)
'''
np.linspace(start,stop,num,endpoint,retstep,dtype)
    start 序列起始值,
    stop  序列终止值,
    num   要生成的等间隔样例的数量,
    endpoint序列中是否包含stop值，默认为ture,
    retstep 若为ture，返回样例(步长),
    dtype  输出ndarray的数据类型
'''
print( np.linspace(1,9,9))                      #从1到9找9个数，包含9
print( np.linspace(1,9,9,endpoint = False))     #从1到9找9个数，不包含9
print(np.linspace(1,9,8,retstep=True))
```

                              深拷贝 与 浅拷贝

1.深拷贝和浅拷贝只针对像Object, Array这样的【引用类型数据】
2.浅拷贝
  是对对象引用地址进行拷贝，并没有开辟新的栈，
  也就是拷贝后的结果是两个对象指向同一个引用地址，
  修改其中一个对象的属性，则另一个对象的属性      也会改变。
3.深拷贝
  则是开启一个新的栈，两个对象对应两个不同的引用地址，
  修改一个对象的属性，    不会改变     另一个对象的属性。

```python
                        # np.fromfunction(函数名,（参数）)
def fun(i,j):
    return (i+1)*(j+1)
a=np.fromfunction(fun,(5,5))    #第一个参数是  函数名 ，第二个参数是数组维度
#同时 第一个 5 传入fun 的 i ，第二个 5 传入fun 的 j
print(a)

print(a[1:3,1:2])    
#定位切片 提取
#     a[ 行位置 , 列位置 ]       （注意：前闭后开）
```

```python
k=np.fromfunction(fun,(5,5))
print(k>10)        # 标记
print(k[k>10])     # 筛选
           
print(np.where(a>10,1,0) ) #条件语句 where 
```

```python
import numpy as np 
        #   axis 指的是数组的轴，二维数组中，axis=0即为列，axis=1即为行
    
        #   在 NumPy中，每一个线性的数组称为是一个轴（axis），也就是维度（dimensions）。
        #   比如说，二维数组相当于是两个一维数组，其中第一个一维数组中每个元素又是一个一维数组。
        #   所以一维数组就是 NumPy 中的轴（axis），第一个轴相当于是底层数组，第二个轴是底层
        #   数组里的数组。而轴的数量——秩，就是数组的维数。
a = np.array([[3,7,5],[8,4,3],[2,4,9]])  
print ('我们的数组是：')
print (a)
print ('\n')
print ('调用 amin() 函数：整个的最小值 ')
print (np.amin(a))             #整个的最小值 
print ('\n')
print ('调用 amin() 函数：每一行的最小值')
print (np.amin(a, axis =1))             #每一行的最小值 
print ('\n')
print ('再次调用 amin() 函数：每一列的最小值')
print (np.amin(a,0))             #每一列的最小值
print ('\n')
print ('调用 amax() 函数：整个的最大值')
print (np.amax(a))             #整个的最大值
print ('\n')
print ('调用 amax() 函数：每一行的最大值')
print (np.amax(a, axis =  1))   #每一行的最大值
print(a.max(axis =  1))                         # 两者等价
print ('\n')
print ('再次调用 amax() 函数：每一列的最大值')
print (np.amax(a,0))    #每一列的最大值
```

```python
# 舍入函数
# numpy.around() 函数返回指定数字的四舍五入值。

# numpy.around(a,decimals)
# 参数说明：

# a: 数组
# decimals: 舍入的小数位数。 默认值为0。 如果为负，整数将四舍五入到小数点左侧的位置
# 实例
import numpy as np
 
a = np.array([1.0,5.55,  123,  0.567,  25.532])  
print  ('原数组：')
print (a)
print ('\n')
print ('舍入后：')
print (np.around(a))
print (np.around(a, decimals =  1))
print (np.around(a, decimals =  -1))
```

```python
# numpy.floor()
# numpy.floor() 返回数字的下舍整数。

# 实例
# import numpy as np
 
a = np.array([-1.7,  1.5,  -0.2,  0.6,  10])
print ('提供的数组：')
print (a)
print ('\n')
print ('修改后的数组：')
print (np.floor(a))
```

```python
# numpy.ceil()
# numpy.ceil() 返回数字的上入整数。

# 实例
import numpy as np
 
a = np.array([-1.7,  1.5,  -0.2,  0.6,  10])  
print  ('提供的数组：')
print (a)
print ('\n')
print ('修改后的数组：')
print (np.ceil(a))
```
