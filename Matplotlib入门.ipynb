# matplotlib

#### 1.单个坐标系显示 ptl.lot（面向过程的画图方法）

```python
# 需求：画出城市11点到12点一小时内，每分钟的温度变化折线图，温度变化在15-18度
                                                # 辅助显示层
import random
import matplotlib.pyplot as plt

from pylab import mpl

# 1.准备数据
x = range(60)
y_shangahi = [random.uniform(15,18) for i in x]

# 2.创建画布
plt.figure(figsize=(20,8),dpi=80)

# 3.绘制图像
plt.plot(x,y_shangahi)

#指定默认字体（这样才能正常显示中文）
mpl.rcParams['font.sans-serif']=['FangSong']

# 修改x，y刻度
# 准备x的刻度说明
x_lable = ["11时{}分".format(i) for i in x]
plt.xticks(x[::5],x_lable[::5])  
plt.yticks(range(0,40,5))
# 添加网格显示
plt.grid(linestyle="--",alpha=0.5)
# 添加描述信息
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("某城市温度变化图")
# 4.显示图像
plt.show()
```

```python
# 需求：再添加一城市温度变化
# 收集到北京当天温度变化，温度在1到3度
                                        # 图像层
import random
import matplotlib.pyplot as plt

# 1.准备数据
x = range(60)
y_shangahi = [random.uniform(15,18) for i in x]
y_beijing = [random.uniform(1,3) for i in x]

# 2.创建画布
plt.figure(figsize=(20,8),dpi=80)

# 3.绘制图像
plt.plot(x,y_shangahi,color="r",linestyle="-",label="上海")
plt.plot(x,y_beijing,color="b",linestyle="-.",label="北京")

# 显示图例
plt.legend()

# 修改x，y刻度
# 准备x的刻度说明
x_lable = ["11时{}分".format(i) for i in x]
plt.xticks(x[::5],x_lable[::5])  
plt.yticks(range(0,40,5))

# 添加网格显示
plt.grid(linestyle="--",alpha=0.5)

# 添加描述信息
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("上海，背景温度变化图")

# 4.显示图像
plt.show()
```

#### 2.多个坐标系显示  ptl.subplots（面向对象的画图方法）

```python
'''
figure, axes = plt.subplots(nrows = 1, ncols = 2, **fig_kw)

returns:
  fig:图对象
  ax:
      设置标题等方法不同
      set_xticks
      set_xlabel
      
注意：  plt.函数名()相当于面向过程的画图方法,
        axes.set()_方法名()相当于面向对象的画图方法。
'''  
```

```python
import random
import matplotlib.pyplot as plt
x = range(10)
y = [random.uniform(1,9) for i in x]
fig,ax=plt.subplots(2,2)
ax[1,1].plot(x,y)
plt.show()
```

#### 3.用法示例

```python
import numpy as np
import matplotlib.pyplot as plt 
#matplotlib的pyplot子库提供了各种绘图API

plt.plot([i for i in range(1,16)],np.random.randn(15))#横纵坐标的值

plt.xlabel("Time(s)") #设置X轴的名字
plt.ylabel("Volt") #设置Y轴的名字
plt.title("PyPlot First Example")#设置图标的标题

plt.ylim(-3,3) #设置Y轴的范围

# 改变字体类型，大小等参数
import matplotlib
matplotlib.rcParams['font.size'] = 20

# plt.legend() #显示图例

# plt.savefig('H:/2.png',dpi=400,bbox_inches='tight')  #保存图标函数
#supported formats: eps, pdf, pgf, png, ps, raw, rgba, svg, svgz

plt.show() #显示出我们创建的所有的绘图对象
#注意：show（）使用时应当在savefig（）之后，否则将图片输出之后在保存就啥也没有了
```

```python
# pyplot并不默认支持中文显示，需要rcParams修改字体来实现
# rcParams的属性：
# ‘font.family’ 用于显示字体的名字
# ‘font.style’ 字体风格，正常’normal’ 或斜体’italic’
# ‘font.size’ 字体大小，整数字号或者’large’   ‘x-small’
# eg:
# import matplotlib
# matplotlib.rcParams[‘font.family’] = ‘STSong’
# matplotlib.rcParams[‘font.size’] = 20
# 设定绘制区域的全部字体变成 华文仿宋，字体大小为20

a=[1,2,6]
b=[3,6,9]
c=[1,5,4]
plt.rcParams['font.family'] = 'STSong'
plt.rcParams['font.size'] = 15
plt.plot(a,b,'b-o'   ,   a,c,'r--.') #支持多个参数，可画多个曲线
# plt.plot(x, y, format_string, **kwargs)   x为x轴数据
```
