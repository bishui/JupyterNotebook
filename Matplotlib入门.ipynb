```python
# 111215-18
                                                # 
import random
import matplotlib.pyplot as plt

from pylab import mpl

# 1.
x = range(60)
y_shangahi = [random.uniform(15,18) for i in x]

# 2.
plt.figure(figsize=(20,8),dpi=80)

# 3.
plt.plot(x,y_shangahi)

#
mpl.rcParams['font.sans-serif']=['FangSong']

# xy
# x
x_lable = ["11{}".format(i) for i in x]
plt.xticks(x[::5],x_lable[::5])  
plt.yticks(range(0,40,5))
# 
plt.grid(linestyle="--",alpha=0.5)
# 
plt.xlabel("")
plt.ylabel("")
plt.title("")
# 4.
plt.show()
```

```python
# 
# 13
                                        # 
import random
import matplotlib.pyplot as plt

# 1.
x = range(60)
y_shangahi = [random.uniform(15,18) for i in x]
y_beijing = [random.uniform(1,3) for i in x]

# 2.
plt.figure(figsize=(20,8),dpi=80)

# 3.
plt.plot(x,y_shangahi,color="r",linestyle="-",label="")
plt.plot(x,y_beijing,color="b",linestyle="-.",label="")

# 
plt.legend()

# xy
# x
x_lable = ["11{}".format(i) for i in x]
plt.xticks(x[::5],x_lable[::5])  
plt.yticks(range(0,40,5))

# 
plt.grid(linestyle="--",alpha=0.5)

# 
plt.xlabel("")
plt.ylabel("")
plt.title("")

# 4.
plt.show()
```
