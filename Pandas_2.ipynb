#### 4.Pandas函数应用

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
ser_obj = pd.Series(np.random.randn(10).cumsum())
print(ser_obj.head())
df_obj = pd.DataFrame(np.random.randn(10, 5), columns =
['a', 'b', 'c', 'd', 'e'])
print(df_obj.head())
df_obj.plot()
plt.show()
```

```python
ser_obj.plot(kind='bar')
df_obj.plot(kind='barh')
plt.show()
```
